```python
rng = np.random.default_rng(seed=1234)
n=1000
```

```python
##Setup model
X = np.hstack([
    np.ones((n,1)),
    rng.random((n,5))
])
beta = np.array([7,-3.3,2.4,1.5,9.7,-10])
err = rng.normal(0,0.1,size = n)
y = X @ beta + err
y
```

```python
##fit model

beta_hat = np.linalg.inv(X.T @ X) @ X.T @ y
print (beta_hat)
print (beta)
```
